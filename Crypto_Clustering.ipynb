```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90, 
    title="Price change % per crypto",
    xlabel="Coin ID",
    ylabel="Price change (%)"
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
scaler.fit(df_market_data)
scaled_data = scaler.transform(df_market_data)

scaled_data
```

```python
# Create a DataFrame with the scaled data
# Copy the crypto names from the original data
# Set the coinid column as index
# Display sample data
scaled_df = pd.DataFrame (scaled_data, columns = df_market_data.columns, index = df_market_data.index)
scaled_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_options = list(range(1,12))
k_options
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for k in k_options:
    kmeans_model = KMeans(n_clusters=k, random_state=1, n_init="auto")
    kmeans_model.fit(scaled_df)
    current_inertia = kmeans_model.inertia_ 
    print(current_inertia)
    inertia_values.append(current_inertia)
    #elbow_inertia[k] = current_inertia

print(inertia_values)
#print(elbow_inertia)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_inertia = {"clusters": k_options, "inertia": inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_inertia)
elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_df.hvplot.line(
    x = "clusters",
    y = "inertia",
    title = "Elbow curve",
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for k would be 5, because is where it occurs an inflexion.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model= KMeans(n_clusters=5, random_state=1, n_init=1)
```

```python
# Fit the K-Means model using the scaled data
model.fit(scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
# Print the resulting array of cluster values.
crypto_clusters = model.predict(scaled_df)
print(crypto_clusters)
```

```python
# Create a copy of the DataFrame
df_market_predictions = scaled_df.copy()
df_market_predictions.head()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_predictions['Predicted Clusters'] = crypto_clusters

# Display sample data
df_market_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
df_market_predictions.hvplot.scatter(
    title='Cryptocurrencies Cluster Predictions with K-Means',
    x='price_change_percentage_24h',
    xlabel='Price Change Percentage 24 Hours',
    y='price_change_percentage_7d',
    ylabel='Price Change Percentage 7 Days',
    by='Predicted Clusters',
    hover_cols='coin_id'
)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_model = PCA(n_components=3, random_state=1)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
crypto_pca_data = pca_model.fit_transform(df_market_predictions)

# View the first five rows of the DataFrame. 
crypto_pca_data[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca_model.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

* Total variance is the sum of variances of all principal components

```python
# Create a new DataFrame with the PCA data.
# Creating a DataFrame with the PCA data
crypto_pca_data_df = pd.DataFrame(crypto_pca_data, columns=["PCA1", "PCA2", "PCA3"], index=scaled_df.index) 

# Copy the crypto names from the original data
# Set the coinid column as index
# Display sample data
crypto_pca_data_df.head()
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_options
```

```python
# Create an empty list to store the inertia values
inertia_values_2 = []
inertia_values_2

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for k in k_options:
    kmeans_model_2 = KMeans(n_clusters=k, random_state=1, n_init="auto")
    kmeans_model_2.fit(crypto_pca_data_df)
    current_inertia = kmeans_model_2.inertia_
    print(current_inertia)
    inertia_values_2.append(current_inertia)
```

```python
# Create a dictionary with the data to plot the Elbow curve
# Create a DataFrame with the data to plot the Elbow curve
elbow_inertia_2 = {"clusters": k_options, "inertia": inertia_values_2}

elbow_df_2 = pd.DataFrame(elbow_inertia_2)
elbow_df_2
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_df_2.hvplot.line(
    x='clusters',
    y='inertia',
    title='Elbow Curve Using PCA Data',
    xticks=k
)
```

* **Question:** What is the best value for `k` when using the PCA data?

  * The best value for k, using PCA data is 4

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Create a copy of the DataFrame with the PCA data
df_market_predictions_pca = crypto_pca_data_df.copy()

# Add a new column to the DataFrame with the predicted clusters
# Display sample data
df_market_predictions_pca['Predicted Clusters PCA'] = crypto_clusters_k4
df_market_predictions_pca.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
df_market_predictions_pca.hvplot.scatter(
    x='PCA1',
    y='PCA2',
    by='Predicted Clusters PCA',
    title='Cryptocurrencies Cluster Predictions K-Means using PCA Data',
    hover_cols='coin_id'
)
```
